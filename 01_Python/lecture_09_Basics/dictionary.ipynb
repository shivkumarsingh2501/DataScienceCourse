```python
#frequency counter
number = [1,2,2,3,3,3,4,4,4,4]

frequecny ={}
for num in number:
    
    if num in frequecny:
        frequecny[num] = frequecny[num] + 1
    else:
        frequecny[num] = 1
print(frequecny)        
```

```python
#counting word frequency in nlp
text = "data science is amazing data science is powerful"
words = text.split()
print(words)
freq = {}
for word in words:
    if word in freq:
        freq[word]+=1
    else:
        freq[word]=1
print(freq)         
```

```python
students = [
("Alice","Math"),
("Bob","Science"),
("Charlie","Math"),
("David","Science"),
("Eva","Math")
]

groups = {}
for name,subject in students:
    if subject not in groups:
        groups[subject] = []
    groups[subject].append(name)
print(groups)
```

```python
numbers = [10,25,30,45,50,60]
filtered = []
for num in numbers:
    if num> 40:
        filtered.append(num)
print(filtered)        
```

```python
numbers = [10,25,30,45,50,60]
filtered = [x for x in numbers if x > 40]
print(filtered)
```
